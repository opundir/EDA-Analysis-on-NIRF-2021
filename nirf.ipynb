```python
import pandas as pd
from tabula import read_pdf
import ssl
```

```python
data=[]
```

```python
#reading the csv file
r = pd.read_csv('EDA.txt',error_bad_lines=False)
r
```

```python
for link in r['link']:
  url = link
  try:
    df=read_pdf(url,pages=1)
  except Exception as e:
    print("Error {}".format(e))
  requires = df[1].iloc[0,1:]
  new_data = requires.values
  data.append(new_data)
```

```python
result = pd.DataFrame(data, columns=['No. of MaleStudents','No. of Female Students','Total Students','Within State (Including male & female)','Outside State(Including male & female)','Outside Country (Including male & female)','Economically Backward (Including male & female)','Socially Challenged (SC+ST+OBC Including male & female)','No. of students receiving full tuition fee reimbursement from the State and Central Government','No. of students receiving full tuition fee reimbursement from Institution Funds', 'No. of students receiving full tuition fee reimbursement from the Private Bodies','No. of students who are not receiving full tuition fee reimbursement'])
```

I.	Using the time series find the age of each college on 01/10/2021.

```python
#first converting date_of-establishment into pandas date and time formate and then finding the age
from datetime import date,datetime
top_15['Date_of_establishment']= pd.to_datetime(top_15['Date_of_establishment'])
def from_dob_to_age(born):
    today = pd.to_datetime('2021-10-01')
    return today.year - born.year - ((today.month, today.day) < (born.month, born.day))
top_15['Age']=top_15['Date_of_establishment'].apply(lambda x: from_dob_to_age(x))
```

```python
top_15
```

II.	Using bar plot show the Age of college on y-axis and name of college on x-axis.            

```python
plt.figure(figsize=(10,5))
plt.bar(top_15['S_name'],top_15['Age'],width=.3)
plt.xlabel('Name of the institute')
plt.ylabel('Years')
plt.title('Bar Plot')
```

```python
plt.figure(figsize=(10,10))
sorted=top_15.sort_values(by='Age',axis=0,ascending=True)
plt.plot(sorted['S_name'],sorted['Age'],marker='o')
plt.xlabel('Name of the institution')
plt.ylabel('Years')
plt.title('Line Plot')
```
